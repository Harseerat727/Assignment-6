```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))
x = np.linspace(-10, 10, 400)

y1 = M * x ** 2  
y2 = M * np.sin(x) 

plt.plot(x, y1, label=f'y = {M} * x^2', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y2, label=f'y = {M} * sin(x)', color='red', linestyle='--', linewidth=2)


plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Plots for y = {M} * x^2 and y = {M} * sin(x)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Subject': ['Math', 'Science', 'History', 'English', 'Art'],
    'Score': [85, 92, 78, 88, 95]
}

df = pd.DataFrame(data)

palette = sns.color_palette("Set2", n_colors=len(df))

plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x='Subject', y='Score', data=df, palette=palette)

for index, value in enumerate(df['Score']):
    bar_plot.text(index, value + 2, str(value), ha='center', va='bottom', fontweight='bold')

plt.title('Scores of Students in Different Subjects', fontsize=16)
plt.xlabel('Subject', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = 12345
np.random.seed(roll_number)

data = np.random.randn(50)

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].plot(np.cumsum(data), color='blue', label='Cumulative Sum')
axs[0].set_title('Cumulative Sum Plot')
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Cumulative Sum')
axs[0].grid(True)
axs[0].legend()

axs[1].scatter(np.arange(50), data + np.random.normal(0, 0.5, 50), color='red', label='Scatter with Noise')
axs[1].set_title('Scatter Plot with Noise')
axs[1].set_xlabel('Index')
axs[1].set_ylabel('Value')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv'
df = pd.read_csv(url)

print(df.columns)  
print(df.head())   

plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='Total Profit', data=df, marker='o', color='green')
plt.title('Total Profit of All Months', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Profit', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df.drop(columns=['Month', 'Total Profit']), markers=True)
plt.title('Product Sales Data', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.grid(True)
plt.xticks(range(len(df['Month'])), df['Month'], rotation=45)
plt.legend(df.drop(columns=['Month', 'Total Profit']).columns, title='Products')
plt.show()

df_melted = pd.melt(df, id_vars=['Month'], var_name='Feature', value_name='Value')

plt.figure(figsize=(14, 8))
sns.barplot(x='Month', y='Value', hue='Feature', data=df_melted)
plt.title('Bar Chart for All Features/Attributes', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Features', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
